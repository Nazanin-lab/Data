```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
file1 = pd.read_table("fdiis.dat", delimiter="\s+", names=["del","x","y"], header=None)

fdiis = pd.DataFrame(file1)
log_fdiis = []
for i in fdiis['y']:
    log_fdiis.append(np.log10(i))
fdiis.insert(2, 'log', log_fdiis)
fdiis
```
